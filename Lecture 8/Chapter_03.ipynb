# Data Cleaning
• Handle Missing Values ( `NaN`)
1. `isnull()/isna()` -  Shows boolean DataFrame, True where NaN
2. `isnull().sum()` - Count of NaNs per column
3. `dropna()` -  Drop rows with any NaN
4. `fillna(value)` -  Fills NaN with a value
5. `ffill()` -  Forward Fill (carry previous value)
6. `bfill()` -  Backward Fill (carry next value)

```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")
df
```

```python
# True for null values (NaN)
df.isnull()

df.isna()
```

```python
df.isnull().sum()    # counts missing vals per column

df.dropna()          # drops rows with missing values

df.dropna(axis = 1)  #  drops columns with missing values

df.fillna(0)       # fills NaN with 0

age_mean = df["age"].mean()  # fills column with mean
df["age"].fillna(age_mean)
```

```python
cleaned_data = df.copy()
age_mean = cleaned_data["age"].mean()
cleaned_data["age"] = cleaned_data["age"].fillna(age_mean)

cleaned_data
```

```python
df.fillna(method = "ffill") # forward fill
df.fillna(method = "bfill") # backward fill
```

# Sorting / Ranking
1. `sort_values()`- Sort values in ascending or descending order.
2. `sort_index()` -  Sort indexes in ascending or descending order.
3. `rank()` - Sort values in ascending or descending order, `method` resolves the ties.

```python
# Sorting - values & index
df2 = df2.fillna(50)
df2.sort_values("Income")                   # sort values in ascending
df2.sort_values("Income", ascending=False)  # sort values in descending
df2.sort_values(["Income", "Age"])          # sorts age,if age same then sorts income

sorted_df = df2.sort_values(["Income", "Age"])
sorted_df
```

```python
sorted_df.sort_index()

# reset
sorted_df.reset_index()
sorted_df.reset_index(drop=True)
```

```python
# Replace the position of tax column
sorted_df[["Id" , "Name" ,"Age" , "Country","Gender","Income","New_Income","Tax" ]]
```

```python
# Ranking
sorted_df["Ranking"] = sorted_df["Income"].rank(ascending=False, method="dense")
sorted_df["Ranking"] = sorted_df["Income"].rank(ascending=False, method="min")
sorted_df["Ranking"] = sorted_df["Income"].rank(ascending=False, method="max")
sorted_df
```
